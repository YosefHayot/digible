## XML example

+ for details about tree traversal and iterators, see https://docs.python.org/2.7/library/xml.etree.elementtree.html

```python
document_tree = ET.parse( './data/mondial_database_less.xml' )
```

```python
type(document_tree.getroot() ) #This method is used to find a specific single element in the document
```

```python
# print names of all countries
for child in document_tree.getroot():   # we are using for in loop to thru elements in tree structure 
    print (child.find('car_code'))    # we are printing child.find('name').text) in this case it is a country. 
# 
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):  #this is for all countries 
    print ('*' + element.find('name').text + ':',) # this print out country with special format
    capitals_string = ''                           #this creates a string countainer for cities in country first empty
    for subelement in element.getiterator('city'):  #this is to get and print the cites 
        capitals_string += subelement.find('name').text + ', ' #this a string aggreator basically list cities followed , 
    print (capitals_string[:-2])  # this is string container and index is meant to reduce some cities. 
```

```python
from xml.etree import ElementTree as ET # from python library import ElementTree
```

```python
tree = ET.parse( './data/mondial_database.xml' )   #this code Parsing an entire document with parse() returns an ElementTree instance
#The tree knows about all of the data in the input document, and the nodes of the tree can be searched or manipulated in place
# this is parsing and returning to us an object. We can do things to object. 



```

```python
type(tree)
```

```python
root = tree.getroot()   # going to root of tree
```

```python
root.tag # lets look at name and tag of root 
```

```python
len(root) # let's look at the root 
```

```python
#let's see how many children the root has 

for child in root:
    print(child.tag) 
            
#the children of tree or root are: country, continent, organization, sea, river, lake, island, mountain, desert, and airport 
```

```python
# here I got child called country and printed country name and population(not latest)
for child in tree.iterfind(".//country"):   #this is for all countries 
    print(child.find('name').text + ':' ,child.find('population').text)
    
   
    
```
